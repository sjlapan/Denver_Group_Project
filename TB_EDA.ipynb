```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tax_and_beer_df= pd.read_csv("resources/brewers_association_data_with_tax_data.csv", index_col= [0])
```

```python
beer_zips= tax_and_beer_df.dropna(how= "all", subset= ["Brewery","Address","City"])
```

```python
beer_zips.head()
```

```python
beer_zips_counts_series= beer_zips["ZIP Code"].value_counts().sort_values(ascending= False)

print("Mean Breweries in a ZIP: "+str(beer_zips_counts_series.mean()))
print("Median Breweries in a ZIP: "+str(beer_zips_counts_series.median()))
print("St. Dev. of Breweries in a ZIP: "+str(beer_zips_counts_series.std()))
print("Max Breweries in a ZIP: "+str(beer_zips_counts_series.max()))
```

```python
beer_zips["AVERAGE TOTAL INCOME"]= beer_zips["TOTAL INCOME AMOUNT"]/beer_zips["NUMBER OF RETURNS WITH TOTAL INCOME"]
```

# Groupby ZIP Code to get each ZIP's average income

```python
brewery_count= beer_zips.groupby("ZIP Code")["ZIP Code"].count()
total_number_of_returns= beer_zips.groupby("ZIP Code")["NUMBER OF RETURNS WITH TOTAL INCOME"].max()
total_income_amount= beer_zips.groupby("ZIP Code")["TOTAL INCOME AMOUNT"].max()
```

```python
print(type(brewery_count))
print(type(total_number_of_returns))
print(type(total_income_amount))
```

```python
beer_zips_average_incomes_df= pd.DataFrame({
    "NUMBER OF BREWERIES IN A ZIP": brewery_count,
    "NUMBER OF RETURNS WITH TOTAL INCOME": total_number_of_returns,
    "TOTAL INCOME AMOUNT": total_income_amount,
    "AVERAGE INCOME AMOUNT": total_income_amount/total_number_of_returns
}, index=brewery_count.index)
```

```python
beer_zips_average_incomes_df.head()
```

# Groupby... again...
## on number of breweries in a ZIP now.

```python
brewery_count_by_zip_aggregations= beer_zips_average_incomes_df.groupby("NUMBER OF BREWERIES IN A ZIP")

by_mean= brewery_count_by_zip_aggregations["TOTAL INCOME AMOUNT"].sum()/ \
    brewery_count_by_zip_aggregations["NUMBER OF RETURNS WITH TOTAL INCOME"].sum()

by_stdev= brewery_count_by_zip_aggregations["TOTAL INCOME AMOUNT"].std()

print(by_mean.tail(10))
print(by_stdev.tail(10))
```

```python
brewery_count_by_zip_aggregations_df= pd.DataFrame({
    "AVERAGE INCOME BY NUMBER OF BREWERIES IN A ZIP": by_mean,
    "STANDARD DEVIATION OF AVERAGE INCOME BY NUMBER OF BREWERIES IN A ZIP": by_stdev
}, index=by_mean.index)

brewery_count_by_zip_aggregations_df.tail()
```

```python
plt.scatter(
    brewery_count_by_zip_aggregations_df.index,
    brewery_count_by_zip_aggregations_df["AVERAGE INCOME BY NUMBER OF BREWERIES IN A ZIP"]
)

plt.title("Average Income by Number of Breweries in a ZIP")
plt.xlabel("Number of Breweries in a ZIP")
plt.ylabel("Average Income ($ in thousands)")

plt.savefig("Images/Average_Income_by_Number_of_Breweries_in_a_ZIP.png")
plt.show()
```
