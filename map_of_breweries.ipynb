```python
import gmaps
import pandas as pd
from laurens_config import api_key
gmaps.configure(api_key=api_key)
```

```python
lat_lngs = pd.read_csv("lats_lngs_breweries.csv")
lat = lat_lngs.lat
list(lat)
lng = lat_lngs.lng
list(lng)
lat_lng_list = list(zip(lat, lng))
lat_lng_list
```

```python
fig = gmaps.figure(layout={"width": "900px", "height": "600px"}, center=(39.83, -96.58), zoom_level=4.4)
symbols = gmaps.symbol_layer(lat_lng_list, fill_color = "green")
fig.add_layer(symbols)
fig
```
