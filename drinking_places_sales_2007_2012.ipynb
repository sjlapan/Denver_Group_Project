```python
import pandas as pd
```

```python
columns = {"GEO.display-label": "State", "ESTAB": "Num_Establishments", "NAICS.display-label": "Category", "RCPTOT": "Sales_1,000", "EMP": "Num_Employees"}
```

```python
industry_data_2007 = pd.read_csv("drinking_places_by_state_2007.csv")
industry_data_2007.head()
```

```python
industry_data_2007_headings = industry_data_2007[["GEO.display-label", "NAICS.display-label", "ESTAB", "RCPTOT", "EMP"]]
industry_data_2007_headings.head()
```

```python
drinking_places_2007 = industry_data_2007_headings.loc[industry_data_2007_headings["NAICS.display-label"] == 
                                                       "Drinking places (alcoholic beverages)"]
drinking_places_2007.head()
```

```python
industry_data_2012 = pd.read_csv("drinking_places_by_state_2012.csv")
industry_data_2012.head()
```

```python
industry_data_2012_headings = industry_data_2012[["GEO.display-label", "NAICS.display-label", "ESTAB", "RCPTOT", "EMP"]]
industry_data_2012_headings.head()
```

```python
drinking_places_2012 = industry_data_2012_headings.loc[industry_data_2012_headings["NAICS.display-label"] == 
                                                       "Drinking places (alcoholic beverages)"]
drinking_places_2012.head()
```

```python
drinking_places_2012_by_state = drinking_places_2012.groupby("GEO.display-label").max()
drinking_places_2012_by_state.head(20)
```
