```python
import pandas as pd
import numpy as np
```

```python
tax_data = pd.read_csv("resources/Tax_DF_Select_Columns_For_Merge.csv", index_col=0)
tax_data.head()
```

```python
tax_data = tax_data.rename(columns = {
    "ZIPCODE": "ZIP Code"
})
tax_data["ZIP Code"] = tax_data["ZIP Code"].astype("str")
```

```python

tax_data.dtypes
```

```python
tax_data.count()
```

```python
beer_data = pd.read_csv("resources/brewers_association_no_duplicates_or_PF.csv", index_col=0)
beer_data.head()
```

```python
beer_data["ZIP Code"] = beer_data["ZIP Code"].astype("str")
beer_data.dtypes
```

```python
beer_data.head()
```

```python
# Gotta ditch those decimals

def trim_fraction(text):
    if ".0" in text:
        return text[:text.rfind(".0")]
    return text
```

```python
beer_data["ZIP Code"] = beer_data["ZIP Code"].apply(trim_fraction)
beer_data.head()
```

```python
full_merge_df = pd.merge(tax_data, beer_data, how = "outer",
                        on= "ZIP Code")
full_merge_df.head()
```

```python
full_merge_df["STATE"].fillna(value= full_merge_df["State"], inplace = True)
full_merge_df.tail()
```

```python
full_merge_df.head()
```

```python
print(len(full_merge_df["STATE"].value_counts()))
```

```python
full_merge_df.drop(["State"], axis = 1, inplace = True)
full_merge_df.head()
```

```python
full_merge_df.to_csv("resources/brewers_association_data_with_tax_data_2.csv")
```
